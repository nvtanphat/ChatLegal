# Trực quan hóa dữ liệu VN Law Chatbot
Notebook này phân tích và trực quan hóa các tập dữ liệu luật và QA trong dự án.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from collections import Counter

# Cấu hình font cho tiếng Việt (nếu cần)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.unicode_minus'] = False
```

## 1. Phân tích Dữ liệu Văn bản Luật (Chunks)

```python
chunks_path = '../data/processed/legal_docs/legal_chunks.json'
with open(chunks_path, 'r', encoding='utf-8') as f:
    chunks_data = json.load(f)

df_chunks = pd.DataFrame(chunks_data)
df_chunks['char_count'] = df_chunks['text'].apply(len)
df_chunks['word_count'] = df_chunks['text'].apply(lambda x: len(x.split()))

print(f"Tổng số chunks: {len(df_chunks)}")
df_chunks.head()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_chunks['char_count'], bins=30, kde=True, color='skyblue')
plt.title('Phân bố độ dài ký tự của Chunks')

plt.subplot(1, 2, 2)
sns.histplot(df_chunks['word_count'], bins=30, kde=True, color='salmon')
plt.title('Phân bố số lượng từ của Chunks')

plt.tight_layout()
plt.show()
```

## 2. Phân tích Nguồn Dữ liệu

```python
plt.figure(figsize=(10, 6))
df_chunks['title'].value_counts().head(10).plot(kind='barh', color='teal')
plt.title('Top 10 Văn bản có nhiều Chunks nhất')
plt.xlabel('Số lượng Chunks')
plt.gca().invert_yaxis()
plt.show()
```

## 3. Phân tích Dữ liệu QA

```python
qa_path = '../data/raw/qa_pairs/qa_pairs_raw.json'
if os.path.exists(qa_path):
    with open(qa_path, 'r', encoding='utf-8') as f:
        qa_data = json.load(f)
    
    # Xử lý tùy thuộc vào cấu trúc file QA của bạn
    # Giả sử là một list các câu hỏi
    df_qa = pd.DataFrame(qa_data)
    print(f"Tổng số câu hỏi QA: {len(df_qa)}")
    
    df_qa['q_len'] = df_qa['question'].apply(len)
    df_qa['a_len'] = df_qa['answer'].apply(len)
    
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.boxplot(y=df_qa['q_len'], color='lightgreen')
    plt.title('Độ dài câu hỏi (ký tự)')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df_qa['a_len'], color='orange')
    plt.title('Độ dài câu trả lời (ký tự)')
    
    plt.tight_layout()
    plt.show()
else:
    print("Không tìm thấy file QA.")
```
